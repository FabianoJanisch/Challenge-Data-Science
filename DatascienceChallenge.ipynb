```python
import pandas as pd
```

```python
data = pd.read_json('https://github.com/sthemonica/alura-voz/blob/main/Dados/Telco-Customer-Churn.json?raw=true')
data
```

```python
data_customer = pd.json_normalize(data['customer'])
data_customer
```

```python
data_phone = pd.json_normalize(data['phone'])
data_phone
```

```python
data_internet = pd.json_normalize(data['internet'])
data_internet
```

```python
data_account = pd.json_normalize(data['account'])
data_account
```

```python
dadosC =  pd.concat([data['customerID'], data['Churn'], data_customer, data_phone, data_internet, data_account], axis=1)
map_senior = {0:'No', 1:'Yes'}
dadosC['SeniorCitizen'].replace(map_senior, inplace=True)
dadosC
```

```python
dadosC.info()
```

Criando funÃ§Ã£o para analisar coluna


```python
def analyse_column(dataframe, column, deleteNa: bool=False):
  value_unique = dataframe[column].unique()
  value_count = dataframe[column].value_counts()
  value_total = dataframe[column].count()

  print (f'Valores Ãºnicos: {value_unique}')
  if '' in value_unique:
    print (f'HÃ¡ valores nulos em {column}')
  print (f'\nValores contÃ¡veis: \n{value_count}')
  print (f'\nValores totais: {value_total}')
```

```python
index = dadosC[dadosC['Churn'] == ''].index
dadosC.drop(index, inplace=True)
```

```python
analyse_column(dadosC, 'Churn')
```
